# Q1 - Camera Projection Model

```python
# Set a focal length (in pixels)
focal_length_mm = 1.5
pixels_mm_x = 0.05
pixels_mm_y = 0.04
focal_x_pixels = focal_length_mm / pixels_mm_x
focal_y_pixels = focal_length_mm / pixels_mm_y
# Set the principal point of the image frame (in pixels) - recall origin is top left
cx = 320
cy = 240
# Define 3D point in camera frame
X_c = 0.5
Y_c = 0.2
Z_c = 2.0
p_c = np.hstack([X_c, Y_c, Z_c])
# Compute (u, v) in the image frame for the point
u_c = cx + focal_x_pixels * X_c / Z_c
v_c = cy + focal_y_pixels * Y_c / Z_c
# Estimate FOV
w_pixels = 640
h_pixels = np.int16((cy * (w_pixels * pixels_mm_x) / cx) / pixels_mm_y)
FOV_x = 2*np.arctan(w_pixels / (2 * focal_x_pixels))
FOV_y = 2*np.arctan(h_pixels / (2 * focal_y_pixels))
print("FOV: ", round(FOV_x, 3), ", ", round(FOV_y, 3))
```
